```python
import os
import requests
import codecs

def convert_hwpx_to_txt(file_path):
    api_endpoint = "https://docsconverter-example.cloud.hancom.com/hwp/doc2txt"
    base_url = "https://docsconverter-example.cloud.hancom.com"
    params = {'file_path': file_path}

    with open(file_path, 'rb') as file:
        response = requests.post(api_endpoint, files={'file': file}, params=params)
    
    if response.status_code == 200:
        response_json = response.json()
        if response_json["docsconverter"]["result"]["code"] == "0000":
            relative_text_file_url = response_json["docsconverter"]["file"]["resource_file"][0]
            text_file_url = f"{base_url}{relative_text_file_url}"
            print(text_file_url)
            
            text_response = requests.get(text_file_url)
            if text_response.status_code == 200:
                content_type = text_response.headers.get('Content-Type')
                encoding = 'utf-8'
                if content_type and 'charset=' in content_type:
                    encoding = content_type.split('charset=')[-1]
                
                print("encoding=", encoding)
                text_response.encoding = encoding
                text_content = text_response.text
                
                # 텍스트 파일로 저장하여 내용 확인
                output_file_path = "converted_text.txt"
                with codecs.open(output_file_path, 'w', encoding='utf-8') as output_file:
                    output_file.write(text_content)
                print(f"Text content saved to: {output_file_path}")

                return text_content
            else:
                print(f"Failed to fetch text content: {text_response.status_code} - {text_response.text}")
        else:
            print(f"Failed to convert file: {response_json['docsconverter']['result']['message']}")
    else:
        print(f"Failed to convert file: {response.status_code} - {response.text}")
```

```python
# 예제 파일 경로 및 출력 디렉토리
input_file = "data/(첨부5) 2024년도 하반기 중남미 지역기구 파견인턴 선발 공고.hwpx"
# 변환 및 텍스트 내용 출력
text_content = convert_hwpx_to_txt(input_file)
if text_content:
    print(f"Text Content:\n{text_content}")
```
